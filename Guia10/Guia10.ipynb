Ejercicio 1

```python
# .1
def contar_lineas(nombre_archivo: str) -> int:
  f = open (nombre_archivo,"r")
  lines: int = len(f.readlines())
  f.close()
  return lines

contar_lineas("/content/drive/MyDrive/Guia_10_archivos/archivo1.txt")
```

```python
# .2
def existe_palabra(palabra: str, nombre_archivo: str) -> bool:
  res: bool = False
  with open(nombre_archivo, "r") as f:
    lines = f.readlines()
    for line in lines:
      words = line.split(" ")
      for word in words:
        if word.strip("\n") == palabra:
          res = True

  return res

print(existe_palabra("son", "/content/drive/MyDrive/Guia_10_archivos/archivo1.txt")) 
```

```python
# .3
def cantidad_apariciones(nombre_archivo: str, palabra: str) -> int:
  res: int = 0
  with open(nombre_archivo, "r") as f:
    lines = f.readlines()
    for line in lines:
      words = line.split(" ")
      for word in words:
        if word.strip("\n") == palabra:
          res += 1

  return res

print(cantidad_apariciones("/content/drive/MyDrive/Guia_10_archivos/archivo1.txt", "son")) 
```

```python
def clonar_sin_comentarios(nombre_archivo: str):
  f_out = open(nombre_archivo.strip(".txt") + "_out.txt","w")
  with open(nombre_archivo,"r") as f_in:
    lines = f_in.readlines()
    for line in lines:
      if not(line.strip(' ').startswith("#")):
        f_out.write(line)
    
  f_out.close()

clonar_sin_comentarios("/content/drive/MyDrive/Guia_10_archivos/archivo2.txt")
```

```python
def clonar_sin_comentarios(nombre_archivo: str):
  f_out = open(nombre_archivo.strip(".txt") + "_out.txt","w")
  with open(nombre_archivo,"r") as f_in:
    lines = f_in.readlines()[::-1]
    lines[0] = lines[0] + "\n"
    lines[-1] = lines[-1].strip("\n")
    f_out.writelines(lines)
    
  f_out.close()

clonar_sin_comentarios("/content/drive/MyDrive/Guia_10_archivos/archivo3.txt")
```
