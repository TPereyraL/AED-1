```python
# .1
def contar_lineas(nombre_archivo: str) -> int:
  f = open (nombre_archivo,"r")
  lines: int = len(f.readlines())
  f.close()
  return lines

contar_lineas("archivo1.txt")
```

```python
# .2
def existe_palabra(palabra: str, nombre_archivo: str) -> bool:
  res: bool = False
  with open(nombre_archivo, "r") as f:
    lines = f.readlines()
    for line in lines:
      words = line.split(" ")
      for word in words:
        if word.strip("\n") == palabra:
          res = True

  return res

print(existe_palabra("son", "archivo1.txt")) 
```

```python
# .3
def cantidad_apariciones(nombre_archivo: str, palabra: str) -> int:
  res: int = 0
  with open(nombre_archivo, "r") as f:
    lines = f.readlines()
    for line in lines:
      words = line.split(" ")
      for word in words:
        if word.strip("\n") == palabra:
          res += 1

  return res

print(cantidad_apariciones("archivo1.txt", "son")) 
```

```python
def clonar_sin_comentarios(nombre_archivo: str):
  f_out = open(nombre_archivo.strip(".txt") + "_out.txt","w")
  with open(nombre_archivo,"r") as f_in:
    lines = f_in.readlines()
    for line in lines:
      if not(line.strip(' ').startswith("#")):
        f_out.write(line)
    
  f_out.close()

clonar_sin_comentarios("archivo2.txt")
```

```python
def clonar_sin_comentarios(nombre_archivo: str):
  f_out = open(nombre_archivo.strip(".txt") + "_out.txt","w")
  with open(nombre_archivo,"r") as f_in:
    lines = f_in.readlines()[::-1]
    lines[0] = lines[0] + "\n"
    lines[-1] = lines[-1].strip("\n")
    f_out.writelines(lines)
    
  f_out.close()

clonar_sin_comentarios("archivo3.txt")
```

Ejercicio 5

```python
import os
import shutil

def insertar_frase(nombre_archivo: str, frase: str):
    temp_file_path = nombre_archivo +'.temp'

    with open(temp_file_path, 'w') as temp_file:
        temp_file.write(frase + '\n')

    with open(nombre_archivo, 'r') as original_file, open(temp_file_path, 'a') as temp_file:
        shutil.copyfileobj(original_file, temp_file)
    
    os.replace(temp_file_path, nombre_archivo)

insertar_frase('archivo5.txt', 'dou\ndou\ndou\ndou')
```

Ejercicio 10

```python
from queue import LifoQueue as pila

def cantidad_elementos(p: pila) -> int:
    c: int = 0
    while not(p.empty()):
        p.get()
        c += 1
    return c

p = pila()
p.put(1)
p.put(2)
p.put(3)

print(cantidad_elementos(p))
```

Ejercicio 11

```python
from queue import LifoQueue as pila

def buscar_maximo(p: pila) -> int:
    max: int = p.get()
    while not(p.empty()):
        tope = p.get()
        if tope > max:
            max = tope
        
    return max

p = pila()
p_list = [4,17,6,18,2,6]

p.queue.extend(p_list)

print(buscar_maximo(p))
```
